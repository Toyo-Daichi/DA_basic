```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./oscillation_KF.csv')
```

```python
df
```

```python
timestep = df[' timestep'].tolist()
```

```python
x_true, x_sim, x_da = df[' x_true'].tolist(), df[' x_sim'].tolist(), df[' x_da'].tolist()
v_true, v_sim, v_da = df[' v_true'].tolist(), df[' v_sim'].tolist(), df[' v_da'].tolist()
```

```python
obs = df.loc[ :, [' timestep', ' obs_data']]
obs = obs[obs[' obs_data'] != 'None']
obs_timestep, obs_score = obs[ ' timestep'].tolist(), obs[ ' obs_data'].tolist()
obs_score = [float(s) for s in obs_score]
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(timestep, x_true, 'b:', label='True')
ax.plot(timestep, x_sim, 'r:', label='Simulation')
ax.plot(timestep, x_da, 'g:', label='Data assimlation')
ax.scatter(obs_timestep, obs_score, marker='*', color='b', label='Observation')
ax.legend()
ax.set_xlabel("time [seconds]")
ax.set_ylabel("damped oscillation [x(m)]")
ax.set_ylim(-5.5, 5.5)
plt.savefig('./tmp_fullx.png')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(timestep[0:30], x_true[0:30], 'b:', label='True')
ax.plot(timestep[0:30], x_sim[0:30], 'r:', label='Simulation')
ax.plot(timestep[0:30], x_da[0:30], 'g:', label='Data assimlation')
ax.scatter(obs_timestep[0:15], obs_score[0:15], marker='*', color='b', label='Observation')
ax.set_xlabel("time [seconds]")
ax.set_ylabel("damped oscillation [x(m)]")
ax.set_ylim(-5.5, 5.5)
plt.savefig('./tmp_shortx.png')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(timestep, v_true, 'b:', label='True')
ax.plot(timestep, v_sim, 'r:', label='Simulation')
ax.plot(timestep, v_da, 'g:', label='Data assimlation')
ax.legend()
ax.set_xlabel("time [seconds]")
ax.set_ylabel("damped oscillation [v(m/s)]")
ax.set_ylim(-3.5, 3.5)
plt.savefig('./tmp_fullv.png')
```
